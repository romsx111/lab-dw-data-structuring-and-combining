In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Upload the datasets

import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
#Standarize header names
def clean_columns (df):
    df.columns = (
        df.columns
        .str.lower()
        .str.replace(" ","_"))
    df = df.rename(columns = {"st":"state"})

    return df

df1 = clean_columns (df1)
df2 = clean_columns (df2)
df3 = clean_columns (df3)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3.columns
```

```python
df = pd.concat([df1, df2, df3], ignore_index = True)

df.head(10)
```

```python
#Clean data
#Drop duplicates

df = df.drop_duplicates()

#Drop null values of entire row

df = df.dropna(how = 'all')

df.head(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['state'].unique()
```

```python
df['gender'].unique()
```

```python
df['education'].unique()
```

```python
df['vehicle_class'].unique()
```

```python
df['number_of_open_complaints'].unique()
```

```python
#Clean gender
df["gender"] = df["gender"].str.lower()

df["gender"] = df["gender"].replace({
    "female": "F",
    "f": "F",
    "femal": "F",
    "male": "M",
    "m": "M",
})

df["gender"] = df["gender"].fillna(df["gender"].mode()[0])
df.isna().sum()
```

```python
df.head(5)
```

```python
#Clean state
df["state"] = df["state"].replace({
    "Cali": "California",
    "AZ": "Arizona",
    "WA": "Washington",
})

df["state"].unique()
```

```python
#Standardized education
df["education"] = df["education"].replace("Bachelors","Bachelor")
df['education'].unique()
```

```python
# Standardized vehicle class
df["vehicle_class"] = df["vehicle_class"].replace({
    "Luxury SUV": "Luxury",
    "Sports Car": "Luxury",
    "Luxury Car": "Luxury",
})

df["vehicle_class"].unique()
```

```python
#Customer lifetime value must be a number not a %

df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(str)
df["customer_lifetime_value"] = df["customer_lifetime_value"].str.replace("%","")
df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(float)

media_clv = df["customer_lifetime_value"].mean()
df["customer_lifetime_value"] = df["customer_lifetime_value"].fillna(media_clv)

df.head(5)
```

```python
df.isna().sum()
```

```python
#Standardized open complaints (must be the middle number)
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.split("/").str[1]
df["number_of_open_complaints"] = df["number_of_open_complaints"].fillna("0").astype(int)
df.head()
```

```python
#Drop rows without customer 

df = df.dropna(subset = ["customer"])
df.isna().sum()
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df.head()
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
import pandas as pd

url="https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

df=pd.read_csv(url)
df.head(7)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
#sales_channel? income - total_claim_amount

pivot_revenue = pd.pivot_table(
    df,
    values = "customer_lifetime_value",
    index = "sales_channel",
    aggfunc="sum")
pivot_revenue = pivot_revenue.round(2) #Round the total revenue to 2 decimal points

pivot_revenue = pivot_revenue.sort_values(by="customer_lifetime_value", ascending = False)

pivot_revenue
    
```

```python
'''The Agent sales channel brought the highest total revenue, being the most effective.

There's potential for optimization in low performing channels such as Call Center and Web. '''
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_education_gender = pd.pivot_table(
    df,
    values = "customer_lifetime_value",
    index = "gender",
    columns = "education",
    aggfunc="mean")
pivot_education_gender = pivot_education_gender.round(2) #Round the total revenue to 2 decimal points

pivot_education_gender
```

```python
'''Customer lifetime value is very similar across genders.

In terms of education, both genders in high school or below show the highest lifetime value, follow by Master's.
The education group with the lowest lifetime value is Doctor.

Education level seems to have more impact on CLV, with lower education levels having higher customer lifetime values.'''
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
